````python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format='retina'

import scipy.stats
from scipy.stats import norm
from scipy.stats import uniform
from scipy import optimize
from astroML import stats as astroMLstats
import time

# MRS Read in the data and store the color info, classifications, and redshifts
data = np.load('../../../homeworks/hw_data/hw8_data_1.npy')
print(data.dtype.names)
u_g = data['u'] - data['g']
g_r = data['g'] - data['r']
r_i = data['r'] - data['i']
i_z = data['i'] - data['z']
z_shift = data['redshift']
specClass = data['specClass']

# MRS Print number of each object type
print('Number of QSOs: ', len(specClass[np.where(specClass == 'QSO')]))
print('Number of Galaxies: ',len(specClass[np.where(specClass == 'GALAXY')]))
````

````python
#M MRS Scatter plot
colors = ["lightblue" if spec == "QSO" else "tomato" for spec in specClass]
plt.scatter(u_g, g_r, c=colors, s=0.5)
# MRS Silly little trick to plot my legend instead of entering legend elements
plt.plot(np.nan, np.nan, 'o', color='tomato', label='Galaxy')
plt.plot(np.nan, np.nan, 'o', color='lightblue', label='Quasar')
plt.xlim(-0.5, 2.5)
plt.ylim(-0.5, 1.5)
plt.xlabel('(u-g)')
plt.ylabel('(g-r)')
plt.show()
````

````python
from  sklearn.model_selection import train_test_split

# MRS stack all of the colors and redshifts
X = np.vstack([u_g, g_r, r_i, i_z, z_shift]).T

# MRS create target labels of 0 and 1 for galaxies and Qso respectively
y = np.array([1 if spec == 'QSO' else 0 for spec in specClass])

# MRS Split the data into test and train datasets
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.2)

from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import roc_curve

# MRS Some parts of code from Lecture 19 Notebook

#----------------------------------------------------------------------
# perform Naive Bayes
for i in range(1, X.shape[1] + 1):

    # MRS Initialize the classifier and train on the 
    clf = GaussianNB()
    clf.fit(X_train[:, :i], y_train)
 
    # MRS Pick out the QSOs from classifier class
    quasar_idx = np.where(clf.classes_ == 1)[0][0]
    probs = clf.predict_proba(X_test[:, :i])[:, quasar_idx]

    # MRSCompute ROC curve 
    fpr, tpr, thresh = roc_curve(y_test, probs)

    # MRS add (0, 0) as first point
    fpr = np.concatenate([[0], fpr])
    tpr = np.concatenate([[0], tpr])
    # Plot the ROC curve
    plt.plot(fpr, tpr, label=f'{i} Feature')

# Final plot tweaks
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Gaussian Naive Bayes ROC Curves')
plt.xlim(0, 0.01) 
plt.ylim(0, 1.05)
plt.legend(loc='lower right')
plt.grid(True)
plt.show()

````

````python
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis as QDA

# MRS Copied from above but now with QDA
for i in range(1, X.shape[1] + 1):
    # MRS Initialize the classifier and train on the 
    clf = QDA()
    clf.fit(X_train[:, :i], y_train)
 
    # MRS Pick out the QSOs from classifier class
    quasar_idx = np.where(clf.classes_ == 1)[0][0]
    probs = clf.predict_proba(X_test[:, :i])[:, quasar_idx]

    # Compute ROC curve 
    fpr, tpr, thresh = roc_curve(y_test, probs)

    # add (0, 0) as first point
    fpr = np.concatenate([[0], fpr])
    tpr = np.concatenate([[0], tpr])
    # Plot the ROC curve
    plt.plot(fpr, tpr, label=f'{i} Feature')

# Final plot tweaks
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('QDA ROC Curves')
plt.xlim(0, 0.01)  # Zoomed in on FPR
plt.ylim(0, 1.05)
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
````

````python
from astroML.classification import GMMBayes

# MRS Remaking data set but with only the 4 colors not redshift
X = np.vstack([u_g, g_r, r_i, i_z]).T
y = np.array([1 if spec == 'QSO' else 0 for spec in specClass])

# MRS split the data into train and test sets
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.2)

# MRS Functionally same code as above but with GMMBayes now 
for n_comps in range(1, 11):
    clf = GMMBayes(n_components=n_comps, covariance_type='full')
    clf.fit(X_train, y_train)
    # MRS Pick out the QSOs from classifier class
    quasar_idx = np.where(clf.classes_ == 1)[0][0]
    probs = clf.predict_proba(X_test)[:, quasar_idx]

    # Compute ROC curve 
    fpr, tpr, thresh = roc_curve(y_test, probs)

 # add (0, 0) as first point
    fpr = np.concatenate([[0], fpr])
    tpr = np.concatenate([[0], tpr])
    # Plot the ROC curve
    plt.plot(fpr, tpr, label=f'{n_comps} Components')

# Final plot tweaks
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('GMMBayes ROC Curves')
plt.xlim(0, 0.01)  # Zoomed in on FPR
plt.ylim(0, 1.05)
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
````

````python
from sklearn.neighbors import KNeighborsClassifier

# MRS Same code as above but now with the KNeighborsClassifier
for n_nbors in range(5, 101, 10):
    clf = KNeighborsClassifier(n_neighbors=n_nbors)
    clf.fit(X_train, y_train)
    # MRS Pick out the QSOs from classifier class
    quasar_idx = np.where(clf.classes_ == 1)[0][0]
    probs = clf.predict_proba(X_test)[:, quasar_idx]

    # Compute ROC curve 
    fpr, tpr, thresh = roc_curve(y_test, probs)

    # add (0, 0) as first point
    fpr = np.concatenate([[0], fpr])
    tpr = np.concatenate([[0], tpr])
    # Plot the ROC curve
    plt.plot(fpr, tpr, label=f'{n_nbors} Neighbors')

# Final plot tweaks
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('K-Nearest Neighbors ROC Curves')
plt.xlim(0, 0.01)  # Zoomed in on FPR
plt.ylim(0, 1.05)
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
````

````python
from sklearn.tree import DecisionTreeClassifier


# MRS Same code as above with the DecisionTreeClassifier
for n_tree in range(1, 11):
    clf = DecisionTreeClassifier(max_depth=n_tree, criterion='entropy')
    clf.fit(X_train, y_train)
    # MRS Pick out the QSOs from classifier class
    quasar_idx = np.where(clf.classes_ == 1)[0][0]
    probs = clf.predict_proba(X_test)[:, quasar_idx]

    # Compute ROC curve 
    fpr, tpr, thresh = roc_curve(y_test, probs)


    # add (0, 0) as first point
    fpr = np.concatenate([[0], fpr])
    tpr = np.concatenate([[0], tpr])
    # Plot the ROC curve
    plt.plot(fpr, tpr, label=f'{n_tree} max depth')

# Final plot tweaks
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Decision Tree Classifier ROC Curves')
plt.xlim(0, 0.01)  # Zoomed in on FPR
plt.ylim(0, 1.05)
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
````
